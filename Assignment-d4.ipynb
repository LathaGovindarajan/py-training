2. Pandas Series and DataFrames
Q1. Create a Pandas Series from a list of student marks [65, 70, 85, 90, 55] with custom index labels as student names.
Calculate the average marks.
Identify students who scored more than 75.
Q2. Create a DataFrame for 5 employees with columns: ['EmpID', 'Name', 'Department', 'Salary'].
Display the data types of all columns.
Add a new column called Bonus as 10% of the Salary.


```python
import pandas as pd
abc=pd.Series([65, 70, 85, 90, 55],index=['ram','lax','lath','john','carol'])
abc
```

```python
print("The average mark scored is :",abc.mean())
```

```python
print("The  students who scored more than 75 is :",abc[abc>75])
```

```python
#q2
data ={
    "EmpID" : ["Latha", "Lux", "Ash"],
    "Department" : ["DE","IT","HR"],
    "Salary" : [50000,70000,90000]
}
df=pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df["Bonus"] = df["Salary"] * 0.10
```

4. Handling Missing Data
Q1. Create a DataFrame with some missing values in columns like 'Name', 'Age', and 'Score'.
Use isnull() and notnull() to identify missing values.
Drop rows with any missing values and show the resulting DataFrame.
Q2. Load or create a DataFrame with missing values.
Fill missing numeric values with the column mean.
Fill missing text values (e.g., in 'City') with the string 'Unknown'.


```python
import numpy as np
data ={
    "Name" :["Latha","John","lux"],
    "Age" :[24,24,27],
    "Score": [np.nan,90,70]
}

df=pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.dropna(axis=0)
```

```python
data1 ={
    "Name" :["Latha","John","lux"],
    "Age" :[24,np.nan,27],
    "City": [np.nan,"Ch","tn"]
}

df2=pd.DataFrame(data1)
df2
```

```python
df2["Age"] = df2["Age"].fillna(df2["Age"].mean())
df2
```

```python
df2=df2.fillna({'City': 'unknown'})
```
